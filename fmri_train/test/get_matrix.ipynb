```python
path="/data06/jinwang/journal_2024/data/processed_data/CC/fmri_fiber_train"
```

```python
import os
import scipy.io as sio
import numpy as np
fmri_path_list = [os.path.join(path, subject) 
                    for subject in sorted(os.listdir(path)) if "CC_7" in subject]
fmri_array=[]

for fmri_path in fmri_path_list:
    f=sio.loadmat(fmri_path)
    fmri_array.append(f['srvf_values'])
fmri_array=np.concatenate(fmri_array,axis=0)
```

```python
fmri_path_list
```

```python
import time
start_time=time.time()
coor=[]
for i in range(fmri_array.shape[0]):
    cof=np.corrcoef(fmri_array[i])
    if i%100==0:
        print(f"processed {i} fmri, time used: {time.time()-start_time}")
    coor.append(cof)

```

```python
coor_all=np.array(coor)
```

```python
coor_all.shape
```

```python
import torch
coor_all_tensor=torch.tensor(coor_all)

torch.save(coor_all_tensor,"coor_all.pt")
```
